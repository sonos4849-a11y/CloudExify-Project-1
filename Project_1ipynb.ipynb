Data visulaiztion


```python
by_product = df.groupby('PRODUCTLINE')['SALES'].sum().sort_values(ascending=False)
print(by_product.head(10))
```

```python
# Bar chart - Sales by Deal Size
sales_by_dealsize = df.groupby('DEALSIZE')['SALES'].sum().sort_values(ascending=False)
sales_by_dealsize.plot(kind='bar', figsize=(10, 6), color='skyblue')
plt.title('Total Sales by Deal Size')
plt.xlabel('Deal Size')
plt.ylabel('Total Sales (Rs)')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# Line plot - Monthly sales trend
df['ORDERDATE'] = pd.to_datetime(df['ORDERDATE'], format='%m-%d-%y')
monthly_sales = df.groupby(df['ORDERDATE'].dt.to_period('M'))['SALES'].sum()
monthly_sales.plot(kind='line', figsize=(12, 6))
plt.title('Monthly Sales Trend')
plt.xlabel('Month')
plt.ylabel('Revenue (Rs)')
plt.tight_layout()
plt.show()
```

Analysis by catagory

```python
by_region = df.groupby('COUNTRY')['SALES'].sum()
by_region.plot(kind='pie', autopct='%1.1f%%', figsize=(8, 8))
plt.title('Sales Distribution by Country')
plt.ylabel('') # Hide the default 'SALES' label on pie chart
plt.show()
```
